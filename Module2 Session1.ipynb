### Top company of every country w.r.t. employees

```python
countries = f500['country'].unique()
countries
```

```python

for i in countries:
    df=f500[f500['country']==i]
    sort_employees= df.sort_values('employees', ascending= False)
    print(i,sort_employees.iloc[0,0])
```

```python
#Top company of every country w.r.t. employees

top_comp = {}                                       #Dictionary to present data

for i in countries:
    df=f500[f500['country']==i]
    sort_employees= df.sort_values('employees', ascending= False)
    top_comp[i] = sort_employees.iloc[0,0]
```

```python
top_comp
```

```python
#Top company of every country w.r.t. employees

top_comp = pd.Series()                                 #Series gives more presentable form of data

for i in countries:
    df=f500[f500['country']==i]
    sort_employees= df.sort_values('employees', ascending= False)
    top_comp[i] = sort_employees.iloc[0,0]
```

```python
top_comp
```

### Conditional ststements in Pandas

```python
f500[(f500['country'] == 'USA') & (f500['sector'] == 'Technology')]            #Tech companies of USA
```

```python
f500[~(f500['country'] == 'USA') & (f500['sector'] == 'Technology')]              #Tech companies other than USA based
```

```python
f500[(f500['country'] == 'USA') & ((f500['sector'] == 'Technology') | (f500['sector'] == 'Financials'))] 
#Tech and Financial companies of USA
# 'and'& has precedence over 'or' |
```
