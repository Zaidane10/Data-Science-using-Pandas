```python
import pandas as pd
```

```python
dataset = pd.read_csv('World_Happiness_2015.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
region = dataset["Region"].unique()
```

```python
region
```

```python
dataset["Region"].value_counts()
```

```python
dataset[dataset["Region"] == "Southern Asia"]
```

```python
dataset[dataset["Region"] == "Southern Asia"]["Happiness Score"].mean()
```

```python
region_mean = pd.Series()

for i in region:
    df = dataset[dataset["Region"] == i]                #fetching region through comparison loop
    region_mean[i] = df["Happiness Score"].mean()
    
region_mean
```

```python
region_grp = dataset.groupby("Region")
```

```python
region_grp
```

```python
region_grp.get_group("Southern Asia")
```

```python
region_mean = pd.Series()

for i in region:
    df = region_grp.get_group(i)                         #fetching individual group of regions
    region_mean[i] = df["Happiness Score"].mean()
    
region_mean
```

```python
region_grp["Happiness Score"]                     #declares series of each group of regions with their happiness scores
```

```python
region_grp["Happiness Score"].mean()
```

```python
region_grp.groups                               #fetches label index of each item of the group
```

```python
region_grp.size()                        #returns no. of rows in the group(like freq table)
```

```python
hs = region_grp["Happiness Score"]
hs
```

#### Aggrigate function: multiple aggrigate functions simultaneously

```python
hs.agg(["mean","max","min","count"])                  #returns agg func results of a Series simultaneously
```

#### Pivot table does the same job at dataset

```python
dataset.pivot_table(index="Region", values="Happiness Score", aggfunc=["mean", "max", "min", "count"], margins= True)

#margins returns total mean of all columns in the pivot table
```
