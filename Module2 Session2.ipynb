### Data Cleaning

```python
import pandas as pd
```

```python
laptops = pd.read_csv('laptops.csv', encoding = 'Windows-1251')
```

```python
laptops
```

```python
laptops.columns
```

```python
laptops.columns = laptops.columns.str.strip().str.replace('(', '').str.replace(')','').str.lower().str.replace(' ','_')\
.str.replace('operating_system','os')       #Replacing Caps with lower case and spaces with _
```

```python
laptops.columns
```

```python
laptops["screen_size"] = laptops["screen_size"].str.strip('"').astype(float) 

#eliminating " sign from inches and typecasting
```

```python
laptops["screen_size"]
```

```python
#laptops = laptops.rename({"screen_size":"screen_size_inches"}, axis=1)

laptops.rename({"screen_size":"screen_size_inches"}, axis=1, inplace= True)   #inplace updates the environment where its working
```

```python
laptops
```

```python
laptops["screen_size_inches"].mean()
```

```python
#Avg Screen size is 15inches
```

```python
laptops["weight"] = laptops["weight"].str.strip("kg").str.strip("kgs").astype(float)
```

```python
laptops.rename({"weight":"weight_kg"}, axis=1, inplace= True)
```

```python
laptops["weight_kg"]
```

```python
laptops
```

```python
laptops["weight_kg"].mean()
```

```python
#Avg weight is around 2kgs
```

```python
laptops["price_euros"] = laptops["price_euros"].str.replace(",",".").astype(float)
```

```python
laptops["price_euros"]
```

```python
laptops["price_euros"].mean()
```

```python
#Avg price is around 1125 euros
```

```python
laptops["ram"] = laptops["ram"].str.strip("GB").astype(float)
```

```python
laptops["ram"]
```

```python
laptops["ram"].mean()
```

```python
#Avg RAM size is 8GB
```

```python
gpu_split = laptops["gpu"].str.split(n=1, expand=True)   #splits from 1st space    #splits in 2 different columns
```

```python
laptops["gpu_company"] = gpu_split[0]
laptops["gpu_model"] = gpu_split[1]
```

```python
laptops
```

```python
laptops["gpu_company"].value_counts()
```

```python
laptops["gpu_company"].value_counts(normalize=True)*100
```

```python
laptops_gpu = laptops[(laptops["gpu_company"] != "Intel") & (laptops["gpu_company"] != "ARM")]
```

```python
laptops_gpu["gpu_company"].value_counts(normalize=True)*100
```

```python
laptops_gpu["gpu_model"].value_counts()
```

```python
cpu_split = laptops["cpu"].str.split(n=1, expand= True)
```

```python
laptops["cpu_company"] = cpu_split[0]
laptops["cpu_model"] = cpu_split[1]
```

```python
laptops
```

```python
laptops["cpu_company"].value_counts(normalize= True)*100
```

```python
laptops["cpu_model"].value_counts()
```

### Hence its prefereable to include following specs
1) Nvidia company gpu
2) GeForce GTX 1050 or GeForce GTX 1060 gpu model
3) 8GB RAM
4) Intel processor with speed > 2GHz
5) ~15inches screen size
6) 2kgs weight
7) Price of 1125 Euros
